After the analysing and cleaning of the data is done, we can build train and test our model.

the target column (dependent variable) is going to be the "Class" column because we want to predict the type of the transaction (fraudent or not).
All the other columns will be our independent variables.

```python
y = df['Class']
X = df.drop(columns=['Class'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

I decided to use a Decision Tree model. The data is already gone through normalization and standardization, Decision Trees can model complex non-linear relationships in the data and sometimes the first algorithm you try works. The Decision Tree algorithm is intuitive and worked this time.

```python
classifier = DecisionTreeClassifier(random_state=42)
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, predictions))
```

As you can see, after building, training and testing the model, The metrics are great. Accuracy of 0.99957. F1-score of 1. Overall great results.

It's important to try to optimize our model as much as we can, no matter the near perfect metrics. Let's see how much the model improves after fine tuning hyperparameters using GridSearchCV.

("GridSearchCV performs an exhaustive search over a specified hyperparameter grid, evaluating all possible hyperparameter combinations using cross-validated performance metrics. It helps you find the best set of hyperparameters for your machine learning model, leading to improved performance.")

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10]
}

classifier = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(classifier, param_grid, cv=5)
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
print(f'Best Hyperparameters: {best_params}')
classifier = DecisionTreeClassifier(random_state=42, **best_params)
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, predictions))
```

As you can see the accuracy got up from 0.99957 to 0.99977. The process bettered the model by 0.0002. The model is working excellent and I didn't really have to optimize it but it was more as an experience with the 'GridSearchCV'.
